```python
import datetime
date_time = datetime.datetime.now()
data ['Age'] = date_time.year - data['Year']
```

```python
data.head()
```

```python
data.drop('Year', axis=1, inplace=True)
```

```python
data.head(1)
```

```python
# Outlier Removal
import seaborn as sns
sns.boxplot(data['Selling_Price'])
```

```python
sorted(data["Selling_Price"],reverse=True)
```

```python
data = data[~(data['Selling_Price']>=33.0) & (data['Selling_Price']<=35.0)]
```

```python
data.shape
```

```python
data['Fuel_Type'].unique()
```

```python
data['Fuel_Type'] = data['Fuel_Type'].map({'Petrol':0,'Diesel':1,'CNG':2})
```

```python
data['Fuel_Type'].unique()
```

```python
data["Seller_Type"].unique()
```

```python
data["Seller_Type"]= data['Seller_Type'].map({'Individual':0,'Dealer':1})
```

```python
data.head(1)
```

```python
data['Transmission']=data['Transmission'].map({'Manual':0,'Automatic':1})
```

```python
data.head()
```

```python
# store Feature Matrix in X and Response Vector in y
x = data.drop(['Car_Name','Selling_Price'],axis=1)
y = data['Selling_Price']
```

```python
x
```

```python
y
```

```python
# Splitting the dataset into the Training set and Test set  
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_test.shape
```

```python
#Import the models
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
# model Training
lr = LinearRegression()
lr.fit(x_train,y_train)
rf = RandomForestRegressor()
rf.fit(x_train,y_train)
gb = GradientBoostingRegressor()
gb.fit(x_train,y_train)
xgb = XGBRegressor()
xgb.fit(x_train,y_train)
```

```python
#Prediction of Test Data
y_pred_lr = lr.predict(x_test)
y_pred_rf = rf.predict(x_test)
y_pred_gb = gb.predict(x_test)
y_pred_xgb = xgb.predict(x_test)
```

```python
# Model Evaluation
from sklearn import metrics
a = metrics.r2_score(y_test,y_pred_lr)
b = metrics.r2_score(y_test,y_pred_rf)
c = metrics.r2_score(y_test,y_pred_gb)
d = metrics.r2_score(y_test,y_pred_xgb)
```

```python
print(a,b,c,d)
```

```python
accuracy =pd.DataFrame({'Model':['Linear Regression','Random Forest','Gradient Boosting','XGBoost'],'Accuracy':[a,b,c,d]})
accuracy
```

```python
sns.barplot(x=accuracy['Model'], y=accuracy['Accuracy'])
```
